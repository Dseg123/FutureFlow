```python
import pandas as pd
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```python
def signal(t):
    r = np.random.rand(len(t))
    print(r)
    return 4 * np.sin(t) + 2 * np.sin(20 * t) + np.sin(100 * t)
```

```python
freq = 200
```

```python
ts = np.arange(0, 5, 1/freq)
print(ts)
```

```python
plt.plot(ts, signal(ts))
```

```python
num_points = 500000
freq = 200
```

```python
ts = np.arange(0, num_points * 1/freq, 1/freq)
```

```python
plt.plot(ts[:100], signal(ts[:100]))
```

```python
us = signal(ts)
vs = signal(ts)
ws = signal(ts)
d = {'u': us, 'v': vs, 'w': ws}
```

```python
df = pd.DataFrame.from_dict(d)
```

```python
print(df)
```

```python
df.to_csv(f'freq_{freq}_data_2.csv')
```

```python
num_points = 500000
num_cycles = 10000
```

```python
u_eps = 0
v_eps = 0.02
w_eps = 0.1
```

```python
u_arr = []
v_arr = []
w_arr = []
```

```python
for i in range(num_points):
    t = i / num_points * (2 * math.pi * num_cycles)
    u = math.sin(t) + u_eps * (random.random() - 0.5)
    v = math.sin(t) + v_eps * (random.random() - 0.5)
    w = math.sin(t) + w_eps * (random.random() - 0.5)
    u_arr.append(u)
    v_arr.append(v)
    w_arr.append(w)
    
    
```

```python
print(u_arr[:100])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(u_arr[:100])
```

```python
coords = [u, v, w]
```

```python
coords = np.array(coords)
```

```python
d = {'u': u_arr, 'v': v_arr, 'w': w_arr}
```

```python
df = pd.DataFrame.from_dict(d)
```

```python
print(df)
```

```python
df.to_csv("sine_data.csv")
```
