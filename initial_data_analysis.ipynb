```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import scipy.io
mat = scipy.io.loadmat('csat_data_6_1_2022.mat')
```

```python
print(mat['T_int'])
```

```python
print(mat.keys())
```

```python
print(mat['u'])
```

```python
u_vals = mat['u']
v_vals = mat['v']
w_vals = mat['w']
```

```python
u_vals = np.array(u_vals)
v_vals = np.array(v_vals)
w_vals = np.array(w_vals)
```

```python
print((len(u_vals) / 10) // 3600)
```

```python
print(u_vals)
```

```python
plt.plot(u_vals, label='u')
plt.plot(v_vals, label='v')
plt.plot(w_vals, label='w')
plt.title("Wind Measurements 5/27/22")
plt.xlabel("Measurement #")
plt.ylabel("Speed (m/s)")
plt.legend()
```

```python
print(np.mean(np.abs(w_vals)))
print(np.mean(np.abs(u_vals)))
print()
print(np.mean(np.abs(v_vals)))
print()
```

```python
plt.plot(np.sqrt(u_vals**2 + v_vals**2 + w_vals**2), label='total magnitude')
plt.plot(np.sqrt(u_vals**2 + v_vals**2), label='horizontal magnitude')

plt.title("Wind Measurements 5/27/22")
plt.xlabel("Measurement #")
plt.ylabel("Speed (m/s)")
plt.legend()
```
