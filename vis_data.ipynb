```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vk_df = pd.read_csv('von_karman_data.csv')
real_df = pd.read_csv('sample_data.csv')
```

```python
vk_df = vk_df.iloc[:len(real_df)]
```

```python
size = len(vk_df['u'])
t = np.arange(0, size*0.1, 0.1) / 3600
```

```python
plt.plot(t, vk_df['u'], label='u')
plt.plot(t, vk_df['v'], label='v')
plt.plot(t, vk_df['w'], label='w')
plt.legend()
plt.xlabel('Time (hrs)')
plt.ylabel('Wind Speed (m/s)')
plt.title('Synthesized von Karman data')
```

```python
plt.plot(t, vk_df['v'], label='synthesized')
plt.plot(t, real_df['v'], label= 'real')
plt.legend()
plt.xlabel('Time (hrs)')
plt.ylabel('Wind Speed (m/s)')
plt.title('Comparison of Real and Synthesized Wind Components')
```

```python
size = 100
plt.plot(t[:size], vk_df['v'].iloc[-size:], label='synthesized')
plt.plot(t[:size], real_df['v'].iloc[-size:], label= 'real')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Wind Speed (m/s)')
plt.title('Comparison of Short-Term Real and Synthesized Wind Components')
```

```python
size = len(real_df['u'])
t = np.arange(0, size*0.1, 0.1) / 3600
```

```python
plt.plot(t, real_df['u'], label='u')
plt.plot(t, real_df['v'], label='v')
plt.plot(t, real_df['w'], label='w')
plt.xlabel('Time (hrs)')
plt.ylabel('Wind Speed (m/s)')
plt.title('Wind Measurements 5/27/22')
plt.legend()
```
