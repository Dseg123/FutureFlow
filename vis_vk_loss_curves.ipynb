```python
def get_train_loss(experiment_path):
    df = pd.read_csv(experiment_path + '/log.csv')
    train_losses = df['train_losses']
    return train_losses.min()

def get_test_loss(experiment_path):
    df = pd.read_csv(experiment_path + '/log.csv')
    test_losses = df['test_losses']
    return test_losses.min()
```

```python
for path in paths:
    print(path + ": " + str(get_train_loss(path)))
```

```python
names = ["Pers", "Lin_reg", "Lin_flat", "Deep", "RNN", "LSTM", "Trans", "ANNMC"]
train_losses = []
test_losses = []
```

```python
for path in paths:
    train_losses.append(get_train_loss(path))
    test_losses.append(get_test_loss(path))
```

```python
x = np.arange(1, 9)
```

```python
plt.bar(x, train_losses, tick_label=names)
plt.rcParams['xtick.labelsize'] = 'small'
# plt.tight_layout()
plt.xlabel("Model Type")
plt.ylabel("MSE Loss")
plt.title("Train Losses of the von Karman Models")
plt.ylim([0.5, 0.9])
```

```python
plt.bar(x, test_losses, tick_label=names)
plt.rcParams['xtick.labelsize'] = 'small'
# plt.tight_layout()
plt.xlabel("Model Type")
plt.ylabel("MSE Loss")
plt.title("Test Losses of the von Karman Models")
plt.ylim([0.5, 0.9])
```
